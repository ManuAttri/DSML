```python
import pandas as pd
#parent class
class utility:
    
    def __init__(self,file_path = r'C:\Users\Manu.Atri\Desktop\Python Assignments\sample_data.csv'):
        try:
            self.df = pd.read_csv(file_path)
        except Exception as e:
            print(e)
                
        
    #function to change the date format
    
    def change_date_format(self):
        try :
            self.df['DATE'] = pd.to_datetime(self.df['DATE'])
            self.df['DATE']= self.df['DATE'].dt.strftime('%y-%m-%d')
            return self.df
        except Exception as e:
            return e
    
    # calculating the yearly average
    
    def avg(self):
        try:
            self.final_df = self.df.groupby(['First Name','Passing Year'])[['Marks Obtained']].mean()
            self.final_df.reset_index(inplace=True)
            self.final_df.rename({'Marks Obtained':'Yearly Average'}, axis = 1, inplace = True )
            self.df = self.df.merge(self.final_df, on = ['First Name', 'Passing Year'], how = 'inner', sort = True)
            return self.df
        except Exception as e:
            return e
    
    #calculating average on the basis of subjects
    
    def subject_avg(self):
        try:
            self.avg()
            self.df_subjectwise = self.df.groupby(['First Name','Subjects'])[['Marks Obtained']].mean()
            self.df_subjectwise.reset_index(inplace = True)
            self.df_subjectwise.rename({'Marks Obtained':'Subject Wise Average'}, axis = 1, inplace = True)
            self.df =  self.df.merge(self.df_subjectwise, on = ['First Name', 'Subjects'], how = 'inner')
            return self.df
        except Exception as e:
            return e

    #function for assigning grades
    
    def grade(self,average):
        try:
            if average >= 90 and average <= 100:
                return 'A'
            elif average >= 80 and average <= 89:
                return 'B'
            elif average >= 70 and average <= 79:
                return 'C'
            elif average >= 60 and average <= 69:
                return 'D'
            elif average >= 50 and average <= 59:
                return 'E'
            else:
                return 'F'
        except Exception as e:
            return e
            
    #function to implement grades according to averages 
    def apply_grade(self):
        try: 
            self.change_date_format()
            self.subject_avg()
            self.df['Grade(based on Yearly Average)']=self.df['Yearly Average'].apply(self.grade)
            return self.df
        except Exception as e:
            return e
    
#child class    
class child_class(utility):
        def create_csv(self):
            try:
                for date in self.df['DATE'].unique():
                    self.df.loc[self.df['DATE'] == date].to_csv(f'{date}.csv.gz', index = False, compression = 'gzip'),self.df.loc[self.df['DATE'] == date].to_csv(f'{date}.csv', index = False)
            except Exception as e:
                return f'An error occured :'+repr(e)
        
    
           
    
    
    
        
```
